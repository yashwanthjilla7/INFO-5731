## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('punkt')

reviews = [
    "This movie is amazing! I loved every moment of it.",
    "The acting was terrible, and the plot was confusing.",
    "I'm not sure how I feel about this film. It had some good parts but also some bad ones.",
    "An absolute masterpiece! Brilliantly directed and well-executed.",
    "I regret watching this. Waste of time and money.",
]

# 1. Bag-of-Words (BoW)
vectorizer = CountVectorizer()
bow_features = vectorizer.fit_transform(reviews)
print("Bag-of-Words Features:")
print(bow_features.toarray())
print("\n")

# 2. TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(reviews)
print("TF-IDF Features:")
print(tfidf_features.toarray())
print("\n")

# 3. Word Embeddings (Word2Vec)
tokenized_reviews = [word_tokenize(review.lower()) for review in reviews]
word2vec_model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=1, workers=4)
word_embedding_features = [word2vec_model.wv[word] for review in tokenized_reviews for word in review]
print("Word Embedding Features:")
print(word_embedding_features)
print("\n")

# 4. N-grams
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))
ngram_features = ngram_vectorizer.fit_transform(reviews)
print("N-gram Features:")
print(ngram_features.toarray())
print("\n")
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
'''
The paper "Feature selection for text classification: A review" by Deng et al. provides an extensive overview of various feature selection methods. One commonly used method is the Chi-squared () test for independence, which can be applied to rank features based on their importance for text classification. Here's how you can use it in Python:
'''

import numpy as np
from sklearn.feature_selection import mutual_info_classif

# Flatten the word embedding features
flat_word_embedding_features = np.array(word_embedding_features).reshape(len(reviews), -1)

# Combine all features into a single array
all_features = np.concatenate((bow_features.toarray(), tfidf_features.toarray(), flat_word_embedding_features, ngram_features.toarray()), axis=1)

# Sample labels for demonstration
labels = np.array([1, 0, 1, 1, 0])  # Replace with your actual labels

# Calculate mutual information scores for each feature
mi_scores = mutual_info_classif(all_features, labels)

# Create a dictionary with feature names and their corresponding mutual information scores
feature_scores = {f"Feature_{i + 1}": mi_score for i, mi_score in enumerate(mi_scores)}

# Rank features based on importance (descending order)
sorted_features = sorted(feature_scores.items(), key=lambda x: x[1], reverse=True)

# Display the ranked features
print("Ranked Features based on Mutual Information:")
for feature, score in sorted_features:
    print(f"{feature}: {score}")
```

Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):
from sentence_transformers import SentenceTransformer, util
import numpy as np
# warnings import filterwarnings
import warnings
warnings.filterwarnings('ignore')

# text data
reviews = [
    "This movie is amazing! I loved every moment of it.",
    "The acting was terrible, and the plot was confusing.",
    "I'm not sure how I feel about this film. It had some good parts but also some bad ones.",
    "An absolute masterpiece! Brilliantly directed and well-executed.",
    "I regret watching this. Waste of time and money.",
]

# Design a query
query = "What are people saying about this movie?"

# Load the BERT model
model = SentenceTransformer('bert-base-uncased')

# Encode the query and text data
query_embedding = model.encode(query, convert_to_tensor=True)
text_embeddings = model.encode(reviews, convert_to_tensor=True)

# Calculate cosine similarity
cosine_similarity = util.pytorch_cos_sim(query_embedding, text_embeddings)

# Convert to numpy array for sorting
similarity_scores = np.array(cosine_similarity.cpu().detach().numpy().flatten())

# Rank documents based on similarity (descending order)
ranked_indices = np.argsort(similarity_scores)[::-1]

# Display the ranked documents
print("Ranked Documents based on Cosine Similarity:")
for rank, index in enumerate(ranked_indices):
    print(f"Rank {rank + 1}: Similarity = {similarity_scores[index]:.4f}, Text: {reviews[index]}")
```
